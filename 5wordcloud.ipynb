```python
# Import the necessary libraries
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Sample from the reviews of a book
reviews = """
I loved the book!
Amazing book!
Compelling reading!
I found this book to be really inspiring.
Long read, nothing special.
well-developed characters, a real page-turner!
Couldn't put it down!
What an incredible book!
I don't understand what all the great reviews are about.
Boring!
I recommended the book to all my friends!
"""

# Create a word cloud 
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(reviews)

# Show the image
plt.figure(figsize=(12, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  
plt.show()
```
