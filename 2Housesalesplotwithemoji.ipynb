```python
import matplotlib.pyplot as plt
import numpy as np

# House Sales Data Plot
#The plot needs to show: house price, distance and direction from city center,agent commision
# It also needs to show the selling agent's company level (Associate, VP, Partner)

#Below we generate the distance and direction from city center,sale price,and agent's level by using random distributions
# We do this for 10 houses
np.random.seed(125)
num_houses = 10
distances = np.random.uniform(0, 30, num_houses)  # distance from city center
prices = np.random.uniform(400, 2000, num_houses) * 1000  # sale price in thousands
directions = np.random.choice(['N', 'S', 'E', 'W'], num_houses)  # direction from city center
agent_levels = np.random.choice([1, 2, 3], num_houses)  # agent's level

#A smiling face emoji is used to convey agent level,agent commision,direction of house from city center.

#The emoji's size corresponds to the level of the selling agent
# 1 corresponds to Associate, 2 to Vice President, 3 to Partner.
def get_emoji_size(level):
    size_map = {1: 250, 2: 380, 3: 700}
    return size_map.get(level, 120)  # Increased size for better visibility

# Function to determine emoji color based on agent commision 
#Lime: Commision 6%, #Pink:Commision: 5%, Yellow: Commision: 4%, White: Commision less than 4%
def get_emoji_color_new(price):
    if price < 600000:
        return 'white'  # Light yellow for $400k-$600k
    elif price < 800000:
        return 'yellow'        # White for $600k-$800k
    elif price < 1000000:
        return 'pink'         # Pink for $800k-$1 million
    else:
        return 'lime'         # Lime for $1 million-$2 million

# Function to rotate the smiley face based on the direction
def rotate_smiley(direction):
    rotation_map = {'N': 0, 'E': 270, 'S': 180, 'W': 90}
    return rotation_map.get(direction, 0)  # default no rotation if direction not found

# Set up the plot 
plt.figure(figsize=(12, 8))

# Plot the house data
for i in range(num_houses):
    plt.scatter(distances[i], prices[i], s=get_emoji_size(agent_levels[i]), c=get_emoji_color_new(prices[i]), 
                marker='o', edgecolors='black', alpha=0.8)
    plt.text(distances[i], prices[i], "", fontsize=agent_levels[i]*10, 
             rotation=rotate_smiley(directions[i]), ha='center', va='center', fontweight='bold')

# Setting plot labels and title
plt.xlabel('Distance from City Center (km)')
plt.ylabel('Sale Price ($)')
plt.title('House Sales Data for 10 Houses')
plt.grid(True)

# Show the updated plot with the new color scheme and larger emojis
plt.show()
```
